# Keras Practice - Feedforward ANN

Reimplementing the Neural Network built in CPEN 502 for solving the Binary XOR Problem.

Feedforward Neural Network with a single hidden layer consisting of 4 Neurons. Weights are updated via backpropagation using the Mean Squared Error loss and Stochastic Gradient Descent with Momentum.

```python
# Imports
%matplotlib inline
import random
import tensorflow.keras.utils
import tensorflow.keras.models as models
from tensorflow.keras.layers import Reshape,Dense,Dropout,Activation,Flatten
from tensorflow.keras.layers import GaussianNoise
from tensorflow.keras.layers import Convolution2D, MaxPooling2D, ZeroPadding2D
from tensorflow.keras.regularizers import *
from tensorflow.keras.optimizers import *
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow.keras
```

```python
# Seed PRNG
random.seed(777)

# Construct Network
model = models.Sequential()
model.add(Dense(4, input_dim=2, activation='relu', name='Hidden_Layer'))
model.add(Dense(1, activation='sigmoid', name='Output_Layer'))
model.compile(loss='mean_squared_error', optimizer=SGD(learning_rate=0.01, momentum=0.9, nesterov=False),metrics=['binary_accuracy'])
model.summary()
```

```python
# Toy Training Data
import numpy as np
data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")
labels = np.array([[0],[1],[1],[0]], "float32")
```

```python
# Train Model
history = model.fit(data, labels, batch_size=1, epochs=500, verbose=2)
```

```python
# Plot training & validation accuracy values
plt.plot(history.history['binary_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper right')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper right')
plt.show()
```

```python
print(model.predict_classes(data))
```
